```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid", context="paper")
```

```python
cpi = pd.read_csv("../data/raw/CPI_20251125-004525.csv")
```

```python
cpi.head()
```

```python
cpi["Month"] = pd.to_datetime(cpi["Month"], format="%YM%m")
cpi.set_index("Month", inplace=True)
```

```python
cpi.head()
```

```python
cpi.info
```

```python
cols = ["Year-on-year inflation (%) Ghana", "Month-on-month inflation (%) Ghana"]

for col in cols:
    cpi[col] = (
        cpi[col]
        .astype(str)        
        .str.replace(",", ".", regex=False)  
        .str.replace("\u2212", "-", regex=False) 
    )
    
    cpi[col] = pd.to_numeric(cpi[col], errors="coerce")
```

```python
gdp = pd.read_csv("../data/raw/Rebased_GDP.csv")
```

```python
gdp.head()
```

```python
gdp.drop(gdp.columns[gdp.columns.str.contains("Unnamed:", case=False)], axis=1, inplace=True)
```

```python
gdp.head()
```

```python
year_raw = gdp["Year"].astype(str).str.strip()

mask = year_raw.str.match(r"^\d{4}\s*Q[1-4]\*?$", na=False)
gdp = gdp[mask].copy()     

year_clean = (
    gdp["Year"]
    .astype(str)
    .str.replace(r"\s+", "", regex=True)  
    .str.replace("*", "", regex=False)   
)

# 3. Build quarterly index and convert to timestamps
q_index = pd.PeriodIndex(year_clean, freq="Q-DEC")
gdp["quarter_date"] = q_index.to_timestamp(how="end")

gdp = gdp.set_index("quarter_date")
```

```python
gdp = gdp.drop(columns=["Year"])

gdp = gdp.rename(columns={
    "Agriculture": "agri_gdp",
    "Industry": "ind_gdp",
    "Services": "svc_gdp",
    "Total GDP (purchaser prices)": "gdp_total",
    "Total Non-Oil GDP (purchaser prices)": "gdp_non_oil",
    "ow Informal GDP (purchaser prices)": "gdp_informal",

    "Agriculture.1": "agri_gdp_yoy",
    "Industry.1": "ind_gdp_yoy",
    "Services.1": "svc_gdp_yoy",
    "Total GDP (purchaser prices).1": "gdp_total_yoy",
    "Total Non-Oil GDP (purchaser prices).1": "gdp_non_oil_yoy",
    "ow Informal GDP (purchaser prices).1": "gdp_informal_yoy",
})
```

```python
gdp.dtypes
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(gdp.index, gdp["Total GDP  (purchaser prices).1"])
ax.set_title("Total GDP  (purchaser prices) Growth Rate")
ax.set_xlabel("Time")
ax.set_ylabel("Total GDP  (purchaser prices)")
plt.tight_layout()
plt.savefig("../figures/TotalGDP.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(gdp.index, gdp["agri_gdp_yoy"])
ax.set_title("Agriculture Growth Rate")
ax.set_xlabel("Time")
ax.set_ylabel("Agriculture")
plt.tight_layout()
plt.savefig("../figures/Agriculture.png")
plt.show()
```

```python
df = pd.read_csv("../data/raw/Time_Series_Monthly.csv")
```

```python
df.head()
```

```python
df["Variables"].unique()
```

```python
df.dtypes
```

```python
macro_raw = df.copy()
month_cols = ["Jan","Feb","Mar","Apr","May","Jun",
              "Jul","Aug","Sep","Oct","Nov","Dec"]

long = macro_raw.melt(
    id_vars=["Year", "Variables"],
    value_vars=month_cols,
    var_name="Month",
    value_name="value_raw"
)
```

```python
print(long.head(10))
print(long["value_raw"].head(10).tolist())
```

```python
s = long["value_raw"].astype(str).str.strip()


s = s.str.replace(",", "", regex=False)

s = s.str.extract(r'([-+]?\d*\.?\d+)')[0]


long["value"] = pd.to_numeric(s, errors="coerce")

long.loc[long["value"] == 0, "value"] = np.nan

print(long[["Year", "Variables", "Month", "value_raw", "value"]].head(15))
```

```python
long["date"] = pd.to_datetime(
    long["Year"].astype(str) + long["Month"],
    format="%Y%b"
)
```

```python
print(long[["Year", "Month", "date"]].drop_duplicates().head(15))
```

```python
macro_ts = (
    long
    .pivot(index="date", columns="Variables", values="value")
    .sort_index()
)

print(macro_ts.dtypes.head())
print(macro_ts.head())
```

```python
macro_ts.isna().sum()
```

```python
macro_ts["Total Liquidity (M2+)(Millions of Ghana Cedis)"].plot()
```

```python

keep_map = {
    "Total Liquidity (M2+)(Millions of Ghana Cedis)": "m2_plus",
    "Broad Money (M2)(Millions of Ghana Cedis)": "m2",
    "Narrow Money (M1)(Millions of Ghana Cedis)": "m1",
    "Private Sector Credit (Millions of Ghana Cedis)": "private_credit",

    "Monetary Policy Rate (%)": "policy_rate",
    "Inter-Bank Weighted Average (%)": "interbank_rate",
    "Average Commercial Banks Lending Rate (%)": "lending_rate",
    "Ghana Reference Rate (%)": "gh_ref_rate",

    "Headline Inflation (%) Year-on-Year": "inflation_headline",
    "Food Inflation (%) Year-on-Year": "inflation_food",
    "Non-Food Inflation (%) Year-on-Year": "inflation_nonfood",
    "Core Inflation (Adjusted for Energy & Utility) (%) Year-on-Year": "inflation_core",

    "Trade Balance (Millions of US$)": "trade_balance_usd",
    "Merchandise Exports (f.o.b) (Millions of US$)": "exports_usd",
    "Merchandise Imports (f.o.b) (Millions of US$)": "imports_usd",
    "Gross International Reserves (Millions of US$)": "gross_reserves_usd",
    "Net International Reserves (Millions of US$)": "net_reserves_usd",

    "Public Debt/GDP (%)": "debt_public_pct_gdp",
    "Domestic Debt/GDP (%)": "debt_domestic_pct_gdp",
    "External Debt/GDP (%)": "debt_external_pct_gdp",
    "Capital Adequacy Ratio (%)": "capital_adequacy",
    "Non Performing Loans  (%)": "npl_ratio",

    "Bank of Ghana Composite Index of Economic Activity (Real Growth)  (%)": "cieg_real_growth",
    "Bank of Ghana Composite Index of Economic Activity (Real)": "cieg_real_level",
}

macro_clean = macro_ts[list(keep_map.keys())].rename(columns=keep_map)

macro_clean = macro_clean.loc["2000":]


for col in ["m2_plus", "m2", "m1", "private_credit",
            "exports_usd", "imports_usd",
            "gross_reserves_usd", "net_reserves_usd"]:
    macro_clean[col] = macro_clean[col] / 1000.0  


macro_clean["m2_plus_yoy"] = macro_clean["m2_plus"].pct_change(12) * 100
macro_clean["private_credit_yoy"] = macro_clean["private_credit"].pct_change(12) * 100

macro_clean = macro_clean.dropna(how="all")
```

```python
fx = pd.read_csv("../data/raw/fx.csv")
```

```python
fx_new_clean = fx_new[fx_new["Mid Rate"] > 0]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(fx_new_clean.index, fx_new_clean["Mid Rate"])
ax.set_title("Ghana: Interbank FX (GHS per USD, post-2007)")
ax.set_xlabel("Year")
ax.set_ylabel("GHS per USD")
plt.tight_layout()
plt.show()
plt.savefig("../figures/interbankfxtimeseries.png")
plt.show()
```

```python
imf["Date_of_arrangement"] = pd.to_datetime(imf["Date_of_arrangement"], dayfirst=True)
imf["Expiration_date"] = pd.to_datetime(imf["Expiration_date_4"], dayfirst=True)
```

```python
imf["Agreed_SDR"] = imf["Agreed"] * 1000
imf["Drawn_SDR"] = imf["Drawn"] * 1000
imf["Outstanding_SDR"] = imf["Outstanding"] * 1000
```
